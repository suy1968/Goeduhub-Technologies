# References:- https://matplotlib.org/stable/tutorials/text/annotations.html

Annotate draws an arrow connecting two points in an axes.

```python
mafia = np.array([1,2,3,4,5,6,7,8])   #mafia=days
tejas = np.array([60,62,61,58,56,57,46,63])  #tejas=speed

plt.figure(figsize=(10,6))
plt.plot(mafia,tejas,linewidth = 4,color='red',linestyle=':') #Changed line style, line color and line width

plt.xlabel('mafia')  #labelling the x-axis, y-axis and title
plt.ylabel('tejas')
plt.title('mafia tejas Measurement')

plt.annotate('start point', xy=(1, 60), xycoords='data',xytext=(30, -70),
textcoords='offset points',arrowprops=dict(arrowstyle= '<|-|>',
color='grey',lw=3.5,ls='--'),horizontalalignment='right', verticalalignment='bottom')  #starting point check

plt.annotate('low point', xy=(7, 46), xycoords='data',xytext=(35,80),
textcoords='offset points',arrowprops=dict(arrowstyle= '<|-|>',
color='grey',lw=3.5,ls='--'),horizontalalignment='right', verticalalignment='bottom') #low point check

plt.annotate('end point', xy=(8, 63), xycoords='data',xytext=(25, -80),
textcoords='offset points',arrowprops=dict(arrowstyle= '<|-|>',
color='grey',lw=3.5,ls='--'),horizontalalignment='right', verticalalignment='bottom') #end point check

plt.show()
```

# 4-  Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x = np.linspace(0,5,90)  #start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(10,6))
plt.plot(x,y,':g', linewidth = 3)
plt.plot(x,z,'--r',linewidth = 3)
plt.xlabel("tejas")
plt.ylabel("kp")
plt.title("tejas/kp function curve")
plt.legend(['sine','cosine']) # legend entries as seperate strings in a list
plt.show()
```

```python
x = np.arange(0,4*np.pi-1,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel("tejas")
plt.ylabel("kp")
plt.title("tejas/kp function curve")
plt.legend(['sine','cosine']) # legend entries as seperate strings in a list
plt.show()
```

# 6- Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

40-60: Below Average

60-80: Average

80-100: Excellent

```python
Marks = np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])

below_avg = Marks[np.logical_and(Marks >= 40,Marks < 60)]

avg = Marks[np.logical_and(Marks >= 60,Marks < 80)]

excell = Marks[np.logical_and(Marks >= 80,Marks < 100)]
```

```python
print(f'No of Below Average students : {below_avg.size}')

print(f'No of Average students : {avg.size}')

print(f'No of Excellent students : {excell.size}')
```

```python
plt.figure(figsize=(12,6))
plt.hist(Marks, bins = [40, 60, 80,100], align='right', color='lightblue', edgecolor='black')
plt.xlabel('vineet')
plt.ylabel('tejas')
plt.title('Histogram of tejas Marks')
plt.show()
```

```python
plt.figure(figsize=(8,8))
vin = teju.sex.value_counts()
plt.pie(vin, labels = ["Male", "Female"], shadow = True, autopct = "%0.0f%%")
plt.title("male / female Proportion")
plt.show()
```

```python
tej = teju[teju["sex"] == "male"]
```

```python
surbhi = teju[teju["sex"] == "female"]
```

```python
plt.figure(figsize=(12,6))

plt.scatter(tej.fare,tej.age,label='Male',color='blue',marker='*')

plt.scatter(surbhi.fare,surbhi.age,label='Female',color='red',marker='+')

plt.xlabel('Fare')

plt.ylabel('Age')

plt.title('Scatter Plot')
plt.legend()
plt.show()
```
